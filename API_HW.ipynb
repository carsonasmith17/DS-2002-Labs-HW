```python
import json
import pandas as pd
import json
import requests
```

```python
apikey="yourkey"

url = "https://yfapi.net/v6/finance/quote"


stock = input()

querystring = {"symbols":stock}

headers = {
    'x-api-key': ''
    }

response = requests.request("GET", url, headers=headers, params=querystring)

stock_json = response.json()

stock_data = {
    "Ticker Name": [stock_json['quoteResponse']['result'][0]["symbol"]],
    "Name": [stock_json['quoteResponse']['result'][0]["longName"]],
    "Price": [stock_json['quoteResponse']['result'][0]["regularMarketPrice"]],
    "52 Week High": [stock_json['quoteResponse']['result'][0]["fiftyTwoWeekHigh"]],
    "52 Week Low": [stock_json['quoteResponse']['result'][0]["fiftyTwoWeekLow"]]
}

print(stock_data)
```

```python
spotify = pd.DataFrame(stock_data)

print(spotify)
```
